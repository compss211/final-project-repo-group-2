```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
item_info = pd.read_csv('../data/meta_data.csv')
review_data = pd.read_csv('../data/raw_review_data.csv')
```

```python
review_data['timestamp'] = pd.to_datetime(review_data['timestamp'], unit='ms')
review_data_20_25 = review_data[(review_data['timestamp'].dt.year >= 2020) & (review_data['timestamp'].dt.year <= 2025)]
```

```python
review_data_20_25.shape
```

```python
joined_df = pd.merge(review_data_20_25, item_info, on='parent_asin', how='left')
```

```python
joined_df = joined_df.drop(columns=['bought_together'])
```
